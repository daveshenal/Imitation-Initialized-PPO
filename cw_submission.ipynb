As autonomous driving technology advances, one of the fundamental challenges remains the development of safe, reliable, and efficient lane keeping systems. **The goal of this project** is to develop an AI driven lane keeping system that enables a self driving car to navigate autonomously while staying centered in its lane, using machine learning techniques and sensor data (such as camera input). Artificial intelligence (AI) has become a cornerstone of autonomous driving research, enabling vehicles to perceive, interpret, and act in real world traffic environments with increasing autonomy. This literature review explores key AI techniques used in lane detection and lane keeping systems, drawing from recent academic research and applications.

##### **A.2.1. AI in Autonomous Vehicles: Overview**

Self driving cars rely on AI to replace or augment human decision making through data driven models trained on sensor inputs like cameras, LiDAR, radar, and GPS. Modern autonomous systems are built upon multiple AI subfields such as machine learning, deep learning, reinforcement learning, and sensor fusion that together enable environment perception, path planning, and motion control. AI enables vehicles to detect objects, recognize lane markings, interpret traffic signals, and make real‑time decisions based on dynamic conditions (Akshitha Karnati, Mehta and Manu Ks, 2022).

A comprehensive review of AI in autonomous vehicles highlights how machine learning models such as convolutional neural networks (CNNs) and recurrent neural networks (RNNs) are central to perception tasks like object and lane detection, while reinforcement learning (RL) and deep learning support adaptive navigation and decision‑making under uncertainty (Devendra Pratap Singh, 2025).

##### **A.2.2. Lane Detection and Steering Control**

Lane detection is a critical perceptual task for lane keeping. Early approaches relied on classical image processing, edge detection and Hough transforms to extract lane markings from camera images. However, such traditional methods struggle with visual noise, shadows, and poor lighting. Modern AI research typically uses deep neural networks for robust lane extraction: CNN architectures are trained end‑to‑end to segment or classify lane boundaries from camera feeds, enabling more accurate and robust detection in diverse conditions (Gajjar, Sanyal and Shah, 2023).

For example, deep learning based lane detection frameworks optimize lane prediction by learning from large datasets, achieving higher accuracy and robustness compared to hand crafted algorithms. These systems process real driving images to extract lane contours and feed this information to downstream control modules that compute steering corrections (Yordanov et al., 2024).

One recent implementation explores a CNN based behavioral model using procedural driving environments to train lane following agents. The system demonstrated significant autonomy levels in both simulated and real driving footage, illustrating the effectiveness of deep features for lane recognition and steering estimation (Yordanov et al., 2024). 

##### **A.2.3. Deep Learning‑Based Control for Lane‑Keeping**

Beyond perception, AI has been used to model the mapping from sensory inputs to driving actions, particularly steering control using deep learning models. Deep learning approaches like transfer learning and fine tuning of pre trained networks (e.g., Xception, VGG, MobileNet) have been applied to predict steering angles directly from images (Singhal et al., 2019; Ngoc et al., 2023). These models learn driving behaviors from labeled human driving data or simulation logs, resulting in continuous steering outputs for lane keeping tasks (Bojarski et al. 2016; Singhal et al., 2019).

Such end‑to‑end models simplify the autonomous pipeline by integrating perception and control into a single deep network, though they require extensive data and careful training to generalize across road conditions.

##### **A.2.4. Reinforcement Learning for Lane‑Keeping and Navigation**

While deep learning excels at perception, reinforcement learning (RL) provides frameworks for learning control policies through interaction with the driving environment. RL agents receive rewards for desirable behavior (e.g., staying centered in a lane) and penalties for undesirable outcomes (e.g., crossing lane boundaries). This allows vehicles to learn optimal steering and navigation strategies by maximizing long term performance metrics.

Recent research has successfully applied deep reinforcement learning (DRL) to train vehicles that perform lane keeping and navigation tasks based on camera inputs. In one study, multiple RL algorithms including DQN, DDQN, A2C, and PPO were evaluated in simulation to teach vehicles how to remain within lane boundaries and follow predefined paths. The research further investigated Sim2Real transfer techniques (e.g., domain randomization and CycleGAN) to improve performance when moving models from simulation into real world testing (Cheng, Lin and Luo, 2025).

Another review of DRL for autonomous driving highlights how RL formulations define state spaces, action spaces, and reward functions that govern safe and efficient maneuvers like lane changes, ramp merges, and platoon coordination on highways. These studies emphasize the flexibility of RL in complex dynamic traffic scenarios where traditional control strategies may fall short (Irshayyid, Chen and Xiong, 2024).

##### **A.2.5. Sensor Fusion and Hybrid Strategies**

Advanced autonomous driving systems often combine multiple AI techniques to overcome the limitations of single modality approaches. Sensor fusion integrates camera vision with radar and LiDAR to enhance environmental perception, improving robustness in adverse weather or lighting conditions. AI models, especially CNNs and machine learning classifiers, process this multimodal data to improve lane and obstacle detection accuracy, feeding high level information into navigation and control systems (Kamesh, 2025).

Hybrid methods that merge classical planning (e.g., real time path planning, search algorithms) with AI driven perception enable vehicles to plan safe trajectories while adapting to dynamic traffic environments. ## **Part B - Compare and Evaluate AI Techniques**

For the self driving lane keeping problem, we have selected three AI techniques that can be applied effectively: Supervised/End-to-End Vision Based Models (CNN, CNN-LSTM, RNN), Reinforcement Learning (PPO or DQN), and a Hybrid Imitation and Reinforcement Learning approach. Below, we will discuss the strengths, weaknesses, advantages, and disadvantages of each, with examples of how each technique can be applied to lane-keeping. we will also compare and evaluate the input data required and the expected output from each technique, before selecting the technique to explore in depth for the prototype implementation in Part C.

### **B.1. Supervised/End-to-End Vision-Based Models (CNN, CNN-LSTM, RNN)**

Vision‑based deep learning models learn a direct mapping from raw camera input to control outputs (e.g., steering). This includes CNNs that process spatial visual features and CNN‑LSTM / RNN architectures that also model temporal patterns across sequential frames. End‑to‑end learning integrates perception and control into a single model pipeline (Luo et al., 2025).

Strengths:
- High pattern extraction capability: CNNs automatically learn robust visual features (lane lines, road edges) from image data, outperforming classical hand‑crafted vision methods (Luo et al., 2025). 
- Temporal coherence: Recurrent models like CNN‑LSTM can capture frame‑to‑frame motion dynamics, which improves steering predictions in continuous driving tasks (Kortli et al., 2022). 
- Fast inference: Once trained, such networks can run efficiently in real time on embedded hardware.

Weaknesses
- Large labeled data requirement: End‑to‑end models require extensive annotated driving datasets (images paired with steering labels), which are costly to collect and curate. 
- Generalization challenges: Models often perform well on training distributions but may fail under lighting/weather variations or unseen road types (Luo et al., 2025).
- No explicit planning: Predicting instantaneous steering does not inherently model longer term navigational decisions or future consequences.

Example Applications
- CNN steering control: A model trained on front camera images predicts steering angles to maintain lane center based on visual cues (Lee and Liu, 2022). 
- CNN‑LSTM for temporal consistency: A CNN extracts features from each frame and a subsequent LSTM models continuity across frames to produce smoother steering control sequences (Kortli et al., 2022). 

Data Requirements
- Input: Sequences of raw camera images (RGB).
- Output: Continuous control signals (e.g., steering angle predictions).

### **B.2. Reinforcement Learning (PPO or DQN)**

Reinforcement learning (RL) trains agents through reward based interaction with an environment rather than fixed labels. Algorithms like Proximal Policy Optimization (PPO) are policy gradient methods designed for continuous control, while Deep Q‑Networks (DQN) learn discrete action policies. In autonomous driving, RL is used to learn control policies that maximize cumulative rewards (e.g., staying centered in the lane).

Strengths
- Adaptive decision‑making: RL can adjust control policies based on real time feedback, making it suitable for dynamic, stochastic tasks like lane‑keeping across diverse road geometries (Delavari, Khan and Kwon, 2025).
- Continuous control (PPO): PPO naturally handles continuous steering actions, which aligns with how vehicles need to modulate steering. 
- Model free learning: RL does not require explicit environmental models, allowing it to learn directly from interaction in simulation or real world.

Weaknesses
- High training cost: RL generally needs many interactions to learn stable policies, which can be expensive or unsafe in real world without simulation (Delavari, Khan and Kwon, 2025). 
- Reward design sensitivity: Performance hinges on well designed reward functions; poorly designed rewards can lead to unintended behaviors. 
- Sim‑to‑Real gap: Models learned in simulators may not transfer seamlessly to real vehicles due to domain shifts (Cheng, Lin and Luo, 2025).

Example Applications
- PPO for lane‑keeping: PPO is used to optimize continuous steering to maximize lane adherence, where rewards approximate how close the vehicle stays to lane center. 
- DQN for discrete steering adjustments: DQN evaluates discrete steering actions and selects the one with highest expected reward for lane maintenance. 

Data Requirements
- Input: State observations (e.g., lane position vectors, extracted features, velocity); sometimes includes raw or pre processed visual input.
- Output: Control policy outputs (continuous steering for PPO or discrete actions for DQN).

### **B.3. Hybrid Imitation + Reinforcement Learning***

This hybrid approach combines supervised learning (imitation learning via CNN feature extraction) with reinforcement learning (PPO) to address weaknesses of each standalone method. In practice, a CNN pre trained on labeled driving data extracts meaningful features from visual input. These feature vectors replace raw images in the RL agent’s observation space, serving as structured perception input for the PPO learner. This technique is akin to hybrid frameworks in research where imitation learning “warm starts” RL, improving convergence and performance.

Strengths
- Faster convergence: Supervised pre training provides a good initialization, reducing RL's cold start exploration phase and speeding up learning (Zhang and Yan, 2025).
- Better generalization: Combined models leverage expert patterns learned via CNN and environment adaptation learned via RL, addressing limitations of pure imitation (Waymo, 2022).
- Structured observations: Feature extractors reduce input dimensionality and noise, enabling RL to focus on relevant driving information.

Weaknesses
- Model complexity: This method requires careful design of interfaces between CNN outputs and RL states, which can complicate training.
- Still resource intensive: Hybrid models may still need extensive training (CNN and RL phases), and require simulation environments for safe learning.
- Proxy reward challenges: Reward design remains critical for PPO effectiveness.

Data Requirements
- Input: Feature vectors from a trained CNN encoder (e.g., lane coordinates, semantic features).
- Output: Continuous steering commands optimized via RL.

### **B.4. Comparative Evaluation**

### **B.5. Justification for Selected Technique: Hybrid Imitation + Reinforcement Learning**

The Hybrid Imitation + Reinforcement Learning approach is selected due to its ability to combine the benefits of supervised learning and reinforcement learning. By leveraging pre trained CNNs for feature extraction, the model benefits from expert knowledge, reducing the need for extensive exploration in the early stages. This significantly accelerates convergence compared to pure RL methods. Additionally, using PPO for adaptive control allows the model to refine its lane keeping behavior in real time based on interaction with the environment. This hybrid approach provides a robust solution for dynamic, real world driving scenarios, offering improved learning efficiency and generalization.

```python
'''
Print details for each datapoint.
    - WP Index
    - Yaw Adj
    - Steer Angle
    - Direction
    - Is Junction 
'''

import carla
import time
import numpy as np
import math
import sys
import random

sys.path.append('C:/CARLA_0.9.15/PythonAPI/carla')
from agents.navigation.global_route_planner import GlobalRoutePlanner


YAW_ADJ_DEGREES = 25  # Maximum yaw angle

def get_angle(car, wp):
    """Calculate angle between car and waypoint"""
    vehicle_pos = car.get_transform()
    car_x = vehicle_pos.location.x
    car_y = vehicle_pos.location.y
    wp_x = wp.transform.location.x
    wp_y = wp.transform.location.y
    
    distance = ((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
    
    if distance == 0:
        x = 0
        y = 0
    else:
        x = (wp_x - car_x) / distance
        y = (wp_y - car_y) / distance
        
    car_vector = vehicle_pos.get_forward_vector()
    degrees = math.degrees(np.arctan2(y, x) - np.arctan2(car_vector.y, car_vector.x))
    
    # Normalize angle to [-180, 180]
    if degrees < -180:
        degrees = degrees + 360
    elif degrees > 180:
        degrees = degrees - 360
    
    return degrees

def get_distant_angle(car, wp_idx, rte, delta):
    """Get general direction at a distance ahead"""
    if wp_idx + delta < len(rte) - 1:
        i = wp_idx + delta
    else:
        i = len(rte) - 1
    
    # Check for intersection within the look forward
    intersection_detected = False
    for x in range(i - wp_idx):
        if rte[wp_idx + x][0].is_junction:
            intersection_detected = True
            intersection_ref = wp_idx + x
            break
    
    if not intersection_detected:
        result = 0
    else:
        angles_planned = []
        all_angles = []
        
        junction_wps = rte[intersection_ref][0].get_junction().get_waypoints(carla.LaneType.Driving)
        for wp in junction_wps:
            angle = int(get_angle(car, wp[1]))
            
            if wp[1].transform.location.distance(rte[intersection_ref][0].transform.location) > 20:
                for i in range(intersection_ref, len(rte) - 1):
                    if wp[1].transform.location.distance(rte[i][0].transform.location) < 10:
                        angles_planned.append(angle)
                    else:
                        all_angles.append(angle)
        
        angles_planned = list(set(angles_planned))
        all_angles = list(set(all_angles))
        alternative_angles = [item for item in all_angles if item not in angles_planned]
        
        if len(alternative_angles) == 0 or len(angles_planned) == 0:
            result = 0
        elif min(angles_planned) < -25 and (min(alternative_angles) > min(angles_planned)):
            result = -1  # Left turn
        elif max(angles_planned) > 25 and (max(alternative_angles) < max(angles_planned)):
            result = 1   # Right turn
        else:
            result = 0
    
    return result

def select_random_route(position, locs, world_map):
    """Select a random route with minimum distance"""
    point_a = position.location
    sampling_resolution = 1
    grp = GlobalRoutePlanner(world_map, sampling_resolution)
    
    min_distance = 100
    route_list = []
    
    for loc in locs:
        cur_route = grp.trace_route(point_a, loc.location)
        if len(cur_route) > min_distance:
            route_list.append(cur_route)
    
    if route_list:
        return random.choice(route_list)
    return None

def cleanup(world):
    """Destroy all actors"""
    for actor in world.get_actors().filter('*vehicle*'):
        actor.destroy()
    for actor in world.get_actors().filter('*sensor*'):
        actor.destroy()

def main():
    # Connect to CARLA
    client = carla.Client('localhost', 2000)
    client.set_timeout(10)
    world = client.get_world()
    
    # Configure world settings
    settings = world.get_settings()
    settings.synchronous_mode = False
    settings.no_rendering_mode = True
    world.apply_settings(settings)
    
    cleanup(world)
    spawn_points = world.get_map().get_spawn_points()
    
    # Test with one route
    start_point = random.choice(spawn_points)
    vehicle_bp = world.get_blueprint_library().filter('*model3*')
    vehicle = world.try_spawn_actor(vehicle_bp[0], start_point)
    
    if vehicle is None:
        print("Failed to spawn vehicle")
        return
    
    time.sleep(2)
    
    # Generate route
    route = select_random_route(start_point, spawn_points, world.get_map())
    
    if route is None:
        print("Failed to generate route")
        cleanup(world)
        return
    
    print(f"Route length: {len(route)} waypoints")
    print("=" * 80)
    print(f"{'WP Index':<10} {'Yaw Adj':<10} {'Steer Angle':<15} {'Direction':<12} {'Is Junction':<12}")
    print("=" * 80)
    
    gen_dir_angle = 0
    
    # Test more waypoints to find junctions
    test_waypoints = min(100, len(route) - 10)
    
    for idx in range(test_waypoints):
        waypoint = route[idx]
        transform = waypoint[0].transform
        vehicle.set_transform(transform)
        time.sleep(0.5)
        
        initial_yaw = waypoint[0].transform.rotation.yaw
        
        # Update general direction angle outside intersections
        if not waypoint[0].is_intersection and not waypoint[0].is_junction:
            gen_dir_angle = get_distant_angle(vehicle, idx, route, 30)
        
        # Check for lane change
        lane_change = False
        if not waypoint[0].is_intersection and not waypoint[0].is_junction:
            if idx < len(route) - 2:
                if route[idx][0].lane_id != route[idx + 1][0].lane_id:
                    lane_change = True
        
        if lane_change:
            if get_angle(vehicle, route[idx + 1][0]) < 0:
                gen_dir_angle = -1
            else:
                gen_dir_angle = 1
        
        # Test with 3 random yaw adjustments per waypoint
        for i in range(3):
            trans = waypoint[0].transform
            angle_adj = random.randrange(-YAW_ADJ_DEGREES, YAW_ADJ_DEGREES, 1)
            trans.rotation.yaw = initial_yaw + angle_adj
            vehicle.set_transform(trans)
            time.sleep(0.3)
            
            if idx + 5 < len(route) - 1:
                predicted_angle = get_angle(vehicle, route[idx + 5][0])
                
                # Determine direction label
                direction = gen_dir_angle  # -1, 0, or 1
                
                is_junction = "Yes" if waypoint[0].is_junction else "No"
                
                # Only print if something interesting (junction or direction change)
                if waypoint[0].is_junction or gen_dir_angle != 0 or abs(predicted_angle) > 30:
                    print(f"{idx:<10} {angle_adj:+4d}°{'':<5} {predicted_angle:+7.1f}°{'':<6} {direction:+2d}{'':<10} {is_junction:<12}")
    
    print("=" * 80)
    print("\nTest complete!")
    print("\nLegend:")
    print("  Direction: -1 = Left turn, 0 = Straight, +1 = Right turn")
    print("  Steer Angle: Angle to waypoint 5 steps ahead")
    print("  Yaw Adj: Random yaw adjustment applied to vehicle")
    
    cleanup(world)


try:
    main()
except KeyboardInterrupt:
    print("\nInterrupted by user")
except Exception as e:
    print(f"Error: {e}")
    import traceback
    traceback.print_exc()
```

```python
'''
Visualize data features with realtime front cam footage to confirm its correction.
'''

import carla
import time
import numpy as np
import math
import sys
import random
import cv2

sys.path.append('C:/CARLA_0.9.15/PythonAPI/carla')
from agents.navigation.global_route_planner import GlobalRoutePlanner


YAW_ADJ_DEGREES = 25  # Maximum yaw angle

# Camera settings
CAMERA_POS_Z = 1.3
CAMERA_POS_X = 1.4
CAM_HEIGHT = 480
CAM_WIDTH = 640
FOV = 90

def camera_callback(image, data_dict):
    """Callback for RGB camera"""
    data_dict['image'] = np.reshape(np.copy(image.raw_data), (image.height, image.width, 4))[:, :, :3]

def get_angle(car, wp):
    """Calculate angle between car and waypoint"""
    vehicle_pos = car.get_transform()
    car_x = vehicle_pos.location.x
    car_y = vehicle_pos.location.y
    wp_x = wp.transform.location.x
    wp_y = wp.transform.location.y
    
    distance = ((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
    
    if distance == 0:
        x = 0
        y = 0
    else:
        x = (wp_x - car_x) / distance
        y = (wp_y - car_y) / distance
        
    car_vector = vehicle_pos.get_forward_vector()
    degrees = math.degrees(np.arctan2(y, x) - np.arctan2(car_vector.y, car_vector.x))
    
    # Normalize angle to [-180, 180]
    if degrees < -180:
        degrees = degrees + 360
    elif degrees > 180:
        degrees = degrees - 360
    
    return degrees

def get_distant_angle(car, wp_idx, rte, delta):
    """Get general direction at a distance ahead"""
    if wp_idx + delta < len(rte) - 1:
        i = wp_idx + delta
    else:
        i = len(rte) - 1
    
    # Check for intersection within the look forward
    intersection_detected = False
    for x in range(i - wp_idx):
        if rte[wp_idx + x][0].is_junction:
            intersection_detected = True
            intersection_ref = wp_idx + x
            break
    
    if not intersection_detected:
        result = 0
    else:
        angles_planned = []
        all_angles = []
        
        junction_wps = rte[intersection_ref][0].get_junction().get_waypoints(carla.LaneType.Driving)
        for wp in junction_wps:
            angle = int(get_angle(car, wp[1]))
            
            if wp[1].transform.location.distance(rte[intersection_ref][0].transform.location) > 20:
                for i in range(intersection_ref, len(rte) - 1):
                    if wp[1].transform.location.distance(rte[i][0].transform.location) < 10:
                        angles_planned.append(angle)
                    else:
                        all_angles.append(angle)
        
        angles_planned = list(set(angles_planned))
        all_angles = list(set(all_angles))
        alternative_angles = [item for item in all_angles if item not in angles_planned]
        
        if len(alternative_angles) == 0 or len(angles_planned) == 0:
            result = 0
        elif min(angles_planned) < -25 and (min(alternative_angles) > min(angles_planned)):
            result = -1  # Left turn
        elif max(angles_planned) > 25 and (max(alternative_angles) < max(angles_planned)):
            result = 1   # Right turn
        else:
            result = 0
    
    return result

def select_random_route(position, locs, world_map):
    """Select a random route with minimum distance"""
    point_a = position.location
    sampling_resolution = 1
    grp = GlobalRoutePlanner(world_map, sampling_resolution)
    
    min_distance = 100
    route_list = []
    
    for loc in locs:
        cur_route = grp.trace_route(point_a, loc.location)
        if len(cur_route) > min_distance:
            route_list.append(cur_route)
    
    if route_list:
        return random.choice(route_list)
    return None

def cleanup(world):
    """Destroy all actors"""
    for actor in world.get_actors().filter('*vehicle*'):
        actor.destroy()
    for actor in world.get_actors().filter('*sensor*'):
        actor.destroy()

def draw_text_on_image(img, lines):
    """Draw multiple lines of text on image"""
    img_copy = img.copy()
    y_offset = 30
    for line in lines:
        cv2.putText(img_copy, line, (10, y_offset), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
        y_offset += 30
    return img_copy

def main():
    # Connect to CARLA
    client = carla.Client('localhost', 2000)
    client.set_timeout(10)
    world = client.get_world()
    
    # Configure world settings
    settings = world.get_settings()
    settings.synchronous_mode = False
    settings.no_rendering_mode = False  # Enable rendering for camera
    world.apply_settings(settings)
    
    cleanup(world)
    spawn_points = world.get_map().get_spawn_points()
    
    # Spawn vehicle
    start_point = random.choice(spawn_points)
    vehicle_bp = world.get_blueprint_library().filter('*model3*')
    vehicle = world.try_spawn_actor(vehicle_bp[0], start_point)
    
    if vehicle is None:
        print("Failed to spawn vehicle")
        return
    
    time.sleep(2)
    
    # Setup RGB Camera
    camera_bp = world.get_blueprint_library().find('sensor.camera.rgb')
    camera_bp.set_attribute('image_size_x', str(CAM_WIDTH))
    camera_bp.set_attribute('image_size_y', str(CAM_HEIGHT))
    camera_bp.set_attribute('fov', str(FOV))
    camera_init_trans = carla.Transform(carla.Location(z=CAMERA_POS_Z, x=CAMERA_POS_X))
    
    camera = world.spawn_actor(camera_bp, camera_init_trans, attach_to=vehicle)
    
    # Camera data dictionary
    camera_data = {'image': np.zeros((CAM_HEIGHT, CAM_WIDTH, 3))}
    camera.listen(lambda image: camera_callback(image, camera_data))
    
    # Wait for first image
    time.sleep(2)
    
    # Generate route
    route = select_random_route(start_point, spawn_points, world.get_map())
    
    if route is None:
        print("Failed to generate route")
        cleanup(world)
        return
    
    print(f"\nRoute generated with {len(route)} waypoints")
    print("=" * 80)
    print("Press 'n' in console to go to next waypoint, or Ctrl+C to quit")
    print("=" * 80)
    
    # Create OpenCV window
    cv2.namedWindow('RGB Camera with Info', cv2.WINDOW_AUTOSIZE)
    
    # Interactive loop
    while True:
        # Choose random waypoint (avoid last 10 waypoints)
        idx = random.randint(0, len(route) - 11)
        waypoint = route[idx]
        
        # Move vehicle to waypoint
        transform = waypoint[0].transform
        vehicle.set_transform(transform)
        time.sleep(1)
        
        initial_yaw = waypoint[0].transform.rotation.yaw
        
        # Calculate general direction angle
        gen_dir_angle = 0
        if not waypoint[0].is_intersection and not waypoint[0].is_junction:
            gen_dir_angle = get_distant_angle(vehicle, idx, route, 30)
        
        # Check for lane change
        lane_change = False
        if not waypoint[0].is_intersection and not waypoint[0].is_junction:
            if idx < len(route) - 2:
                if route[idx][0].lane_id != route[idx + 1][0].lane_id:
                    lane_change = True
        
        if lane_change:
            if get_angle(vehicle, route[idx + 1][0]) < 0:
                gen_dir_angle = -1
            else:
                gen_dir_angle = 1
        
        # Apply random yaw adjustment
        angle_adj = random.randrange(-YAW_ADJ_DEGREES, YAW_ADJ_DEGREES, 1)
        trans = waypoint[0].transform
        trans.rotation.yaw = initial_yaw + angle_adj
        vehicle.set_transform(trans)
        time.sleep(1)
        
        # Calculate steering angle
        predicted_angle = get_angle(vehicle, route[idx + 5][0])
        
        # Get waypoint info
        is_junction = "Yes" if waypoint[0].is_junction else "No"
        direction_label = {-1: "LEFT TURN", 0: "STRAIGHT", 1: "RIGHT TURN"}[gen_dir_angle]
        
        # Print to console
        print(f"\n--- Waypoint {idx} / {len(route)} ---")
        print(f"Initial Yaw:     {initial_yaw:.1f}°")
        print(f"Yaw Adjustment:  {angle_adj:+d}°")
        print(f"Final Yaw:       {initial_yaw + angle_adj:.1f}°")
        print(f"Steering Angle:  {predicted_angle:+.1f}°")
        print(f"Direction:       {gen_dir_angle:+d} ({direction_label})")
        print(f"Is Junction:     {is_junction}")
        print(f"Lane Change:     {'Yes' if lane_change else 'No'}")
        
        # Display image with text overlay
        while True:
            if np.sum(camera_data['image']) > 0:
                # Prepare text lines
                text_lines = [
                    f"Waypoint: {idx}/{len(route)}",
                    f"Initial Yaw: {initial_yaw:.1f}",
                    f"Yaw Adj: {angle_adj:+d}",
                    f"Steer Angle: {predicted_angle:+.1f}",
                    f"Direction: {gen_dir_angle} ({direction_label})",
                    f"Junction: {is_junction}",
                    f"Lane Change: {'Yes' if lane_change else 'No'}",
                    "",
                    "Press 'n' for next waypoint"
                ]
                
                # Draw text on image
                display_img = draw_text_on_image(camera_data['image'], text_lines)
                cv2.imshow('RGB Camera with Info', display_img)
                
                # Wait for key press
                key = cv2.waitKey(100)
                if key == ord('n') or key == ord('N'):
                    break
            else:
                time.sleep(0.1)
    
    cv2.destroyAllWindows()
    cleanup(world)


try:
    main()
except KeyboardInterrupt:
    print("\n\nInterrupted by user")
    cv2.destroyAllWindows()
except Exception as e:
    print(f"Error: {e}")
    import traceback
    traceback.print_exc()
    cv2.destroyAllWindows()
```

##### 3.1.3. Data Generation

```python
''' 
Generate Data

Image generation rate ~ 2500/Hr
Space per semantic image ~ 35KB
Space per rgb image ~ 440KB

'''

import os
import carla
import time
import cv2
import numpy as np
import math
import sys
import random

sys.path.append('C:/CARLA_0.9.15/PythonAPI/carla') # Carla Path !!
from agents.navigation.global_route_planner import GlobalRoutePlanner


# Create imgs folders if it doesn't exist
os.makedirs('imgs/semantic', exist_ok=True)
os.makedirs('imgs/rgb', exist_ok=True)

SHOW_RGB = False # To display cam footage on a window (Turned off for better performance)

# Cam position relative to the vehicle
CAMERA_POS_Z = 1.3
CAMERA_POS_X = 1.4

# Image size
CAM_HEIGHT = 480
CAM_WIDTH = 640

FOV = 90

YAW_ADJ_DEGREES = 25 # Maximum yaw angle

def camera_callback(image,data_dict):
    data_dict['image'] = np.reshape(np.copy(image.raw_data),(image.height,image.width,4))[:, :, :3]

def sem_callback(image,data_dict):
    #change for Semantic camera
    image.convert(carla.ColorConverter.CityScapesPalette)
    data_dict['sem_image'] = np.reshape(np.copy(image.raw_data),(image.height,image.width,4))[:, :, :3]

def cleanup(): # Destroy all the actors in the simulation
    for actor in world.get_actors().filter('*vehicle*'):
        actor.destroy()
    for actor in world.get_actors().filter('*sensor*'):
        actor.destroy()


# Angle between the car and waypoint
def get_angle(car,wp):
    vehicle_pos = car.get_transform()
    car_x = vehicle_pos.location.x
    car_y = vehicle_pos.location.y
    wp_x = wp.transform.location.x
    wp_y = wp.transform.location.y
    
    if ((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5 == 0:
        x=0
        y=0
    else:
        x = (wp_x - car_x)/((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
        y = (wp_y - car_y)/((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
        
    car_vector = vehicle_pos.get_forward_vector()
    degrees = math.degrees(np.arctan2(y, x) - np.arctan2(car_vector.y, car_vector.x))
    # extra checks on predicted angle when values close to 360 degrees are returned
    if degrees<-180:
        degrees = degrees + 360
    elif degrees > 180:
        degrees = degrees - 360
    return degrees

# get angle to a waypoint at a distance
def get_distant_angle(car,wp_idx,rte, delta):
    if wp_idx + delta < len(rte)-1:
        i = wp_idx + delta
    else:
        i = len(rte)-1
    # check for intersection within the look forward
    # so we do not give turn results when just following the road
    intersection_detected = False
    for x in range(i-wp_idx):
        if rte[wp_idx+x][0].is_junction:
             intersection_detected = True
             intersection_ref = wp_idx+x
             break
    if not intersection_detected:
        result = 0
    else: # check out the intersection
        angles_planned = [] # list of angles towards current exit from the intersection
        all_angles = []
        
        junction_wps = rte[intersection_ref][0].get_junction().get_waypoints(carla.LaneType.Driving)
        for wp in junction_wps:
            angle = int(get_angle(car,wp[1])) 

            if wp[1].transform.location.distance(route[intersection_ref][0].transform.location) > 20: 
  
                for i in range(intersection_ref,len(route)-1):
                    if wp[1].transform.location.distance(route[i][0].transform.location) < 10:
                        angles_planned.append(angle)
                    else:
                        all_angles.append(angle)
        angles_planned = list(set(angles_planned))
        all_angles = list(set(all_angles))
        alternative_angles = [item for item in all_angles if item not in angles_planned] 
        if len(alternative_angles) == 0 or len(angles_planned) == 0:
            result = 0
        elif min(angles_planned)<-25 and (min(alternative_angles) > min(angles_planned)):
            #we are planning left turn
            result = -1
        elif max(angles_planned)>25 and (max(alternative_angles) < max(angles_planned)):
            result = 1
        else:
            result = 0  
    return result


def select_random_route(position,locs):
    point_a = position.location #start at where the car is or last waypoint
    sampling_resolution = 1
    grp = GlobalRoutePlanner(world.get_map(), sampling_resolution)
    # pick the longest possible route
    min_distance = 100
    result_route = None
    route_list = []
    for loc in locs: 
        cur_route = grp.trace_route(point_a, loc.location)
        if len(cur_route) > min_distance:
            route_list.append(cur_route)
    result_route = random.choice(route_list)
    return result_route

def main():
    client = carla.Client('localhost', 2000)
    time.sleep(5) # just in case
    client.set_timeout(10) # increase if failed to connect in time or check for port 2000 
    world = client.get_world()
    settings = world.get_settings()
    settings.synchronous_mode = False
    settings.no_rendering_mode = True
    if settings.synchronous_mode:
        settings.fixed_delta_seconds = 0.05
    world.apply_settings(settings)

    cleanup()
    spawn_points = world.get_map().get_spawn_points()

    while True:
        start_point = random.choice(spawn_points)
        vehicle_bp = world.get_blueprint_library().filter('*model3*')
        vehicle = world.try_spawn_actor(vehicle_bp[0], start_point)
        time.sleep(2)
    
        # RGB CAM
        camera_bp = world.get_blueprint_library().find('sensor.camera.rgb')
        camera_bp.set_attribute('image_size_x', str(CAM_WIDTH))
        camera_bp.set_attribute('image_size_y', str(CAM_HEIGHT))
        camera_bp.set_attribute('fov', str(FOV))
        camera_init_trans = carla.Transform(carla.Location(z=CAMERA_POS_Z,x=CAMERA_POS_X))

        camera = world.spawn_actor(camera_bp,camera_init_trans,attach_to=vehicle)
        image_w = camera_bp.get_attribute('image_size_x').as_int()
        image_h = camera_bp.get_attribute('image_size_y').as_int()
        camera_data = {'image': np.zeros((image_h,image_w,3)),
                    'sem_image': np.zeros((image_h,image_w,3))}
        
        camera.listen(lambda image: camera_callback(image,camera_data))

        # Semantic cam
        sem_camera_bp = world.get_blueprint_library().find('sensor.camera.semantic_segmentation')
        sem_camera_bp.set_attribute('image_size_x', str(CAM_WIDTH))
        sem_camera_bp.set_attribute('image_size_y', str(CAM_HEIGHT))
        sem_camera_bp.set_attribute('fov', str(FOV))
    
        sem_camera = world.spawn_actor(sem_camera_bp,camera_init_trans,attach_to=vehicle)
        
        sem_camera.listen(lambda image: sem_callback(image,camera_data))

        if SHOW_RGB:
            cv2.namedWindow('RGB Camera',cv2.WINDOW_AUTOSIZE)
            cv2.imshow('RGB Camera',camera_data['image'])
        
        route = select_random_route(start_point,spawn_points)
        gen_dir_angle = 0 # in case we did not get a general GPS direction
        for idx, waypoint in enumerate(route): # move the car through the route (Not real navigation !!!)
            
            transform = waypoint[0].transform
            vehicle.set_transform(transform)
            vehicle.apply_control(carla.VehicleControl(throttle=0, steer=0, brake=1))
            time.sleep(2)
            initial_yaw = waypoint[0].transform.rotation.yaw
            # GPS general direction is only taken outside intersections
            if not waypoint[0].is_intersection and not waypoint[0].is_junction:
                # general angle taken before spinning the car
                gen_dir_angle = get_distant_angle(vehicle,idx,route,30)
            # logic to detect a lane change and ignore/not take those images
            lane_change = False
            if not waypoint[0].is_intersection and not waypoint[0].is_junction:
                if idx < len(route)-2:
                    if route[idx][0].lane_id != route[idx+1][0].lane_id:
                        lane_change = True
            if lane_change: # lane changes are treated as turns in general direction
                if get_angle(vehicle,route[idx+1][0])<0:
                    gen_dir_angle = -1
                else:
                    gen_dir_angle = 1

            for i in range(5):
                trans = waypoint[0].transform
                angle_adj = random.randrange(-YAW_ADJ_DEGREES, YAW_ADJ_DEGREES, 1)
                trans.rotation.yaw = initial_yaw +angle_adj 
                vehicle.set_transform(trans)
                vehicle.apply_control(carla.VehicleControl(throttle=0, steer=0, brake=1))
                time.sleep(1) 
                
                
                if SHOW_RGB:
                    cv2.imshow('RGB Camera',camera_data['image'])
                
                if idx +5 < len(route)-1: # Check for end of the route
                    # we always get the angle to +5 waypoint ahead of us
                    predicted_angle = get_angle(vehicle,route[idx+5][0])
                    
                    time_grab = time.time_ns()
                    rgb_image = camera_data['image']
                    sem_image = camera_data['sem_image']
                    if np.sum(sem_image) > 0:
                        cv2.imwrite('imgs/semantic/%019d_%s_%s.png' % (time_grab, gen_dir_angle,
                                                                       round(predicted_angle,0)), sem_image)
                        cv2.imwrite('imgs/rgb/%019d_%s_%s.png' % (time_grab, gen_dir_angle,
                                                                  round(predicted_angle,0)), rgb_image)
    cleanup()
    cv2.destroyAllWindows()


try:
    main()
except KeyboardInterrupt:
    print("\n\nInterrupted by user")
    cv2.destroyAllWindows()
except Exception as e:
    print(f"Error: {e}")
    import traceback
    traceback.print_exc()
    cv2.destroyAllWindows()
```

```python
import cv2
import os
import random
from matplotlib import pyplot as plt


# Constants for cropping
HEIGHT = 240
WIDTH = 320
HEIGHT_REQUIRED_PORTION = 0.5  # Bottom share (e.g., 0.1 means the lowest 10% of rows)
WIDTH_REQUIRED_PORTION = 0.9  # 90% of the width

# Calculate crop parameters for the bottom part of the image
height_from = int(HEIGHT * (1 - HEIGHT_REQUIRED_PORTION))  # Start cropping from this height
width_from = int((WIDTH - WIDTH * WIDTH_REQUIRED_PORTION) / 2)  # Central crop start
width_to = width_from + int(WIDTH_REQUIRED_PORTION * WIDTH)  # End crop position

# Path to image directory
data_dir = 'imgs/semantic/'

# Function to crop the image as per the given parameters
def crop_image(image):
    # Resize to a smaller starting size before cropping
    image = cv2.resize(image, (WIDTH, HEIGHT))

    # Crop the image from the bottom portion and center along width
    cropped_image = image[height_from:, width_from:width_to]
    
    return cropped_image

# Get a list of image file paths
image_files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith('.png')]

# Randomly select 9 images
random_images = random.sample(image_files, 9)

# Set up the plot for displaying 9 images
fig, axs = plt.subplots(3, 3, figsize=(12, 7))
fig.suptitle('Sample Cropped Images', fontsize=16)

# Loop through each subplot and display a random cropped image
for i, ax in enumerate(axs.flat):
    # Load the image
    img_path = random_images[i]
    image = cv2.imread(img_path)

    # Crop the image
    cropped_image = crop_image(image)

    # Convert from BGR (OpenCV format) to RGB (matplotlib format)
    cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

    # Display the image
    ax.imshow(cropped_image_rgb)
    ax.axis('off')  # Turn off axis
    ax.set_title(f"Image {i+1}")

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig('analysis_plots/sample_cropped_images.png', dpi=150, bbox_inches='tight')
print("Sample images saved as 'sample_cropped_images.png'")
plt.show()
```

#### **4.1 Custom Data Generator for CNN Model**

```python
import os
import cv2
import numpy as np
import random


SOURCE_IMG_HIGHT = 480
SOURCE_IMG_WIDTH = 640

# scaled down image size before cropping
HEIGHT = 240
WIDTH = 320

HEIGHT_REQUIRED_PORTION = 0.5 # bottom share, e.g. 0.1 is take lowest 10% of rows
WIDTH_REQUIRED_PORTION = 0.9

MAX_STEER_DEGREES = 40

data_dir = 'imgs/semantic/'

#image re-size and crop calcs
height_from = int(HEIGHT * (1 -HEIGHT_REQUIRED_PORTION))
width_from = int((WIDTH - WIDTH * WIDTH_REQUIRED_PORTION) / 2)
width_to = width_from + int(WIDTH_REQUIRED_PORTION * WIDTH)

# image size into model
new_height = HEIGHT - height_from # = 120
new_width = width_to - width_from # = 288

# Define the parameters for image preprocessing and augmentation
batch_size = 64
image_size = (new_width,new_height)

# Define the position of the label in the image file name
label_position = -5  # label is the fifth character from the end

# Create a custom data generator
def custom_data_generator(image_files, batch_size):
    num_samples = len(image_files)
    while True:
        indices = np.random.randint(0, num_samples, batch_size)
        batch_images = []
        batch_input_2 = []
        batch_labels = []
        for idx in indices:
            image_path = image_files[idx]
            label = float(os.path.basename(image_path).split('.png')[0].split('_')[2])
            if label > MAX_STEER_DEGREES:
                label = MAX_STEER_DEGREES
            elif label < -MAX_STEER_DEGREES:
                label = -MAX_STEER_DEGREES
            label = float(label)/MAX_STEER_DEGREES
            input_2 = int(os.path.basename(image_path).split('.png')[0].split('_')[1])
            image = preprocess_image(image_path)
            batch_images.append(image)
            batch_input_2.append(input_2)
            batch_labels.append(label)
        yield [np.array(batch_images), np.array(batch_input_2)], np.array(batch_labels)

# Preprocess the image
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    # resize to a smaller starting size before cropping
    image = cv2.resize(image, (WIDTH,HEIGHT))
    
    #cropping application to take lower side of the image
    image = image[height_from:,width_from:width_to]
    image = image / 255.0  # Normalize pixel values between 0 and 1
    return image

# Get a list of image file paths and labels
image_files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith('.png')]
random.shuffle(image_files)

# Split the data into training and validation sets
split_index = int(len(image_files) * 0.8)  # 80% for training, 20% for validation
train_files, val_files = image_files[:split_index], image_files[split_index:]

# Create data generators for training and validation
train_generator = custom_data_generator(train_files, batch_size)
val_generator = custom_data_generator(val_files, batch_size)
```

#### **4.1. Model Architecture**

```python
from tensorflow import keras
from keras.models import Model
from keras.layers import Dense, Input, Dropout, MaxPooling2D, Conv2D, Concatenate, Flatten, Activation
from keras import regularizers
from keras.utils import plot_model


def create_model():
    # Image input
    image_input = Input(shape=(new_height, new_width, 3))
    # Integer input
    integer_input = Input(shape=(1,))
    # Preprocess the image input
    x = Conv2D(64, kernel_size=(6, 6), activation='relu',padding='same')(image_input)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, kernel_size=(6, 6), activation='relu',padding='same')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, kernel_size=(6, 6), activation='relu',padding='same')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, kernel_size=(6, 6), activation='relu',padding='same')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Dense(8, activation='relu',activity_regularizer=regularizers.L2(1e-5))(x)
    x = Dropout(0.2)(x)
    x = Dense(4, activation='relu',activity_regularizer=regularizers.L2(1e-5))(x)
    x = Flatten()(x)
    # Concatenate image features with integer input
    concatenated_inputs = Concatenate()([x, integer_input])
    # Dense layers for prediction
    output = Dense(1, activation='linear')(concatenated_inputs)
    # Create the model
    model = Model(inputs=[image_input, integer_input], outputs=output)
    return model

model = create_model()
model.compile(loss='MSE', optimizer='adam')
model.summary()

# Save the model architecture as a PNG image
plot_model(model, to_file='analysis_plots/model_architecture.png', show_shapes=True, show_layer_names=True)
print("Model architecture saved as 'model_architecture.png'")
```

```python
from keras.callbacks import EarlyStopping


# Define EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_files) // batch_size,
    epochs=20,
    validation_data=val_generator,
    validation_steps=len(val_files) // batch_size,
    callbacks=[early_stopping]
)
```

```python
import sys
import random

sys.path.append('C:/CARLA_0.9.15/PythonAPI/carla')
from agents.navigation.global_route_planner import GlobalRoutePlanner
from agents.navigation.local_planner import RoadOption


def path_planner(spawn_points, map):
    """
    Get the longest route waypoints from a starting point to various spawn points.

    Args:
        start_point (carla.Location): The starting location from which to trace routes.
        spawn_points (list[carla.SpawnPoint]): A list of spawn points to which routes will be traced.
        map (carla.Map): The Carla map object, used for route planning.

    Returns:
        list[tuple]: A list of waypoints representing the longest route found,
                     where each waypoint is a tuple containing a Carla waypoint object
                     and its associated location.
    """
    
    def has_lane_change_gap(route, gap_threshold=4):
          
        for i in range(len(route) - 1):
            wp1 = route[i][0]
            wp2 = route[i+1][0]
            
            # 2D distance only (ignoring Z)
            dx = wp2.transform.location.x - wp1.transform.location.x
            dy = wp2.transform.location.y - wp1.transform.location.y
            distance = math.sqrt(dx*dx + dy*dy)
            
            # If distance exceeds threshold, it's likely a lane change
            if distance > gap_threshold:
                return True
                
        return False
       
    sampling_resolution = 1
    grp = GlobalRoutePlanner(map, sampling_resolution)
    
    # Option 1: Random
    spawn_point = random.choice(spawn_points)
    
    # Option 2: Only Intersections
    # intersections = (0, 99, 21, 59, 79, 51, 52, 102, 11, 12, 25, 24)
    # spawn_point = spawn_points[random.choice(intersections)]
    
    # For Option 1 and 2:
    # Pick the longest possible route
    distance = 0
    for loc in spawn_points:
        
        cur_route = grp.trace_route(spawn_point.location, loc.location)
        
        # Remove first and last points from the route to avoide lane changes in last or first waypoint (common)
        cur_route = cur_route[1:-1]
        
        if has_lane_change_gap(cur_route):
            continue
        
        if len(cur_route) > distance:
            distance = len(cur_route)
            route = cur_route
            
    # Option 3: Known start and end point route        
    # spawn_point = spawn_points[8]
    # end_point = spawn_points[87]        
    # route = grp.trace_route(spawn_point.location, end_point.location)
    # route = route[1:-1]
    
    return route, spawn_point


def get_path_waypoints(route):
    """
    Extract the juctions start and end waypoints.

    Args:
        route (list[tuple]): A list of waypoints, where each waypoint is represented 
                             as a tuple containing a Carla waypoint object and its location.

    Returns:
        list[carla.Waypoint]: A list of waypoints
    """
    # List to hold waypoints of different types
    path_waypoints = []
    
    # Flag to track if the last waypoint is a junction
    last_wp_is_junction = route[-1][0].is_junction if route else False

    # Iterate over waypoints in the route
    for i in range(len(route)):
        if route[i][0].is_junction:
            # Check if it's the start of a junction waypoints set
            if (i == 0 or not route[i-1][0].is_junction):  # Start of a junction
                path_waypoints.append(route[i][0])
            # Check if it's the end of a 'j' set
            elif (i == len(route)-1 or not route[i+1][0].is_junction):  # End of a junction
                path_waypoints.append(route[i][0])
                
    # Add the last waypoint if it is not a junction
    if not last_wp_is_junction:
        path_waypoints.append(route[-1][0])

    
    return path_waypoints  # Return the list of waypoints


def update_waypoint(path_waypoints, current_wp_indx, current_loc):
    
    """
    Update to the next waypoint if the vehicle is close enough to the current one.
    
    returns,
        current_wp: next waypoint to reach. (not update if its the destination)
        wp_type: way point type (normal, destination, intersection, other)
        reached: boolean value to know that waypoint has reached
        dist = distance to new waypoint
    
    """
    threshold = 3 # 3m
    current_wp = path_waypoints[current_wp_indx]
    wp_type = 0 # normal by default
    reached = False # current waypoint reached
    destination = False # end of the path
            
    # Calculate distance to the current waypoint
    dist = current_loc.distance(current_wp.transform.location)
    
    # Check if the vehicle is within the threshold distance to update to the next waypoint
    if dist < threshold:
        reached = True
        if current_wp_indx < len(path_waypoints) - 1:
            current_wp_indx += 1  # Move to the next waypoint
            current_wp = path_waypoints[current_wp_indx]  # Update current waypoint
            dist = current_loc.distance(current_wp.transform.location)  # Calculate new distance
            if current_wp_indx == len(path_waypoints) - 1:
                wp_type = 1 # next waypoint is destination
        else:
            destination = True  # Current waypoint is the destination
    return current_wp, current_wp_indx, wp_type, reached, dist, destination


def get_dis_ang_from_nearest_wp(current_loc, vehicle_yaw, route):
    """
    Calculate the distance and angle difference between the vehicle and the nearest waypoint in the route.

    Args:
        vehicle (carla.Vehicle): The vehicle object.
        route (list[tuple]): A list of waypoints from the route, where each entry is a tuple
                             containing a CARLA Waypoint object and a RoadOption.

    Returns:
        tuple: A tuple containing the distance to the nearest waypoint and the absolute angular difference in yaw.
    """
    # Get the vehicle's current location
    selected_wp, dist_to_wp = get_closest_wp(current_loc, route)

    # Check if we found a valid waypoint
    if selected_wp:        
        dir_differ = get_angle(vehicle_yaw, selected_wp.transform.rotation.yaw)

        return dist_to_wp, dir_differ
    
    else:
        print("No waypoint found.")
        return 0.0, 0.0
    

def get_angle(vehicle_yaw, wp_yaw):    
    # Calculate the yaw difference
    dir_differ = vehicle_yaw - wp_yaw

    # Normalize to -180 to 180 degrees
    dir_differ = (dir_differ + 180) % 360 - 180
    return dir_differ

    
def get_closest_wp(current_loc, route):
    min_dist = float('inf')  # Initialize to a very large distance
    closest_wp = None  # To store the closest waypoint
    # Find the closest waypoint in the route
    for wp, _ in route:
        wp_loc = wp.transform.location
        dist = current_loc.distance(wp_loc)
        
        # Update if this waypoint is closer
        if dist < min_dist:
            min_dist = dist
            closest_wp = wp  # Store the closest waypoint object
            
    return closest_wp, min_dist


def find_longest_non_junction_paths(carla_map, top_n=20, step=1.0):
    """
    Find the longest paths without intersections, ensuring no path overlaps with longer paths.
    
    Args:
        carla_map (carla.Map): The CARLA map
        top_n (int): Number of longest paths to return
        step (float): Distance between waypoints when exploring
        
    Returns:
        list: Top N longest non-overlapping paths as tuples (start_wp, end_wp, distance, waypoint_list)
    """
    topology = carla_map.get_topology()
    
    # Step 1: Build a graph of road segments (non-junction to non-junction)
    # We need to find continuous segments between junctions
    print("Building road segment graph...")
    
    all_segments = []
    processed_starts = set()
    
    # Get all waypoints from topology and explore from each
    for wp_start, wp_end in topology:
        # Try to find the beginning of a non-junction segment
        # Start from wp_start and go backwards until we hit a junction or dead end
        start_candidates = [wp_start]
        
        # Also check previous waypoints
        current = wp_start
        for _ in range(100):  # Limit backwards search
            prev_wps = current.previous(step)
            if not prev_wps or prev_wps[0].is_junction:
                break
            current = prev_wps[0]
            start_candidates.append(current)
        
        # Now explore forward from each candidate
        for start_wp in start_candidates:
            if start_wp.is_junction:
                continue
            
            # Create unique ID for this starting point
            start_id = (start_wp.road_id, start_wp.lane_id, 
                       round(start_wp.transform.location.x, 1), 
                       round(start_wp.transform.location.y, 1))
            
            if start_id in processed_starts:
                continue
            
            processed_starts.add(start_id)
            
            # Explore forward
            current_wp = start_wp
            total_distance = 0.0
            path_waypoints = [start_wp]
            
            while True:
                next_wps = current_wp.next(step)
                
                if not next_wps:
                    break
                
                next_wp = next_wps[0]
                
                # Stop if we reach a junction
                if next_wp.is_junction:
                    break
                
                # Calculate distance
                distance_increment = current_wp.transform.location.distance(
                    next_wp.transform.location
                )
                total_distance += distance_increment
                
                # Safety check
                if total_distance >= 200.0:  # or whatever max you want
                    break
                
                path_waypoints.append(next_wp)
                current_wp = next_wp
                
                # Safety check
                if len(path_waypoints) > 10000:
                    print(f"Warning: Path too long, breaking at {total_distance}m")
                    break
            
            # Save segment if it has meaningful length
            if total_distance > step:
                end_wp = path_waypoints[-1]
                all_segments.append((start_wp, end_wp, total_distance, path_waypoints))
    
    print(f"Found {len(all_segments)} total road segments")
    
    # Step 2: Sort by distance (longest first)
    all_segments.sort(key=lambda x: x[2], reverse=True)
    
    # Step 3: Select non-overlapping segments using a spatial approach
    selected_paths = []
    used_locations = set()
    
    for start_wp, end_wp, distance, waypoint_list in all_segments:
        # Check if this path overlaps with any already selected path
        path_overlaps = False
        
        # Sample waypoints along the path to check for overlaps
        sample_indices = list(range(0, len(waypoint_list), max(1, len(waypoint_list) // 10)))
        if len(waypoint_list) - 1 not in sample_indices:
            sample_indices.append(len(waypoint_list) - 1)
        
        path_locations = []
        for i in sample_indices:
            wp = waypoint_list[i]
            loc_id = (wp.road_id, wp.lane_id)
            path_locations.append(loc_id)
            
            if loc_id in used_locations:
                path_overlaps = True
                break
        
        # If no overlap, add this path
        if not path_overlaps:
            selected_paths.append((start_wp, end_wp, distance, waypoint_list))
            
            # Mark all locations in this path as used
            for wp in waypoint_list:
                loc_id = (wp.road_id, wp.lane_id)
                used_locations.add(loc_id)
            
            if len(selected_paths) >= top_n:
                break
    
    print(f"\nSelected {len(selected_paths)} non-overlapping paths:\n")
    
    # for i, (start_wp, end_wp, distance, wps) in enumerate(selected_paths, 1):
    #     print(f"{i}. Distance: {distance:.2f}m, Waypoints: {len(wps)}")
    #     print(f"   Start: Road {start_wp.road_id}, Lane {start_wp.lane_id}, Loc: ({start_wp.transform.location.x:.1f}, {start_wp.transform.location.y:.1f})")
    #     print(f"   End:   Road {end_wp.road_id}, Lane {end_wp.lane_id}, Loc: ({end_wp.transform.location.x:.1f}, {end_wp.transform.location.y:.1f})")
    #     print()
    
    # Convert waypoint lists to formatted routes before returning
    formatted_paths = []
    for _, _, _, waypoint_list in selected_paths:
        # Convert to (waypoint, RoadOption) format
        formatted_route = [(wp, RoadOption.LANEFOLLOW) for wp in waypoint_list]
        formatted_paths.append(formatted_route)
    
    return formatted_paths
```

#### **5.2 Reward and Penalty Calculation**

```python
import carla


def calculate_distance_to_waypoint_reward(reached, current_dist, start_dist):
    """
    Calculate the reward based on the distance to the waypoint and stopping at the destination.
    
    args:
        done: waypoint reched
        current_dist: distance to waypoint
        start_dist: start_distance form the waypoint

   """ 
    # Reward scaling based on distance (closer is better)
    if reached:
        reward = 300.0  # Reward for getting very close to the waypoint
    else:
        reward = max(0.0, 10.0 - (current_dist / start_dist * 10.0))  # Scale reward
    
    return reward


def calculate_speed_reward(done, PREFERRED_SPEED, s):
    dif = abs(PREFERRED_SPEED - s)
    prnalty = 0
    if dif < 5:
        prnalty = 5 - dif
    elif s > PREFERRED_SPEED:
        done = True
        prnalty = -300
    else:           
        prnalty = -dif
    
    return prnalty, done


def collision_penalty(done, collision_hist):
    penalty = 1
    if len(collision_hist) != 0:
        done = True
        penalty = -320
        # print('c')
        return penalty, done
    return penalty, done


def lane_invasion_penalty(done, lane_invade_hist):
    """
    Calculate a penalty for lane invasions based on crossed lane markings.
    Stops checking as soon as an invalid marking is detected.
    """
    penalty = 1
    
    # Define restricted lane markings
    restricted_markings = {
        carla.LaneMarkingType.Solid,
        carla.LaneMarkingType.SolidSolid,
    }
    
    soft_restricted_markings = {
        carla.LaneMarkingType.Broken,
    }
    
    
    # Iterate through each LaneInvasionEvent in the history
    for event in lane_invade_hist:
        # Iterate through the list of crossed lane markings
        for marking in event.crossed_lane_markings:
            # Check if the marking type is in the restricted markings set
            if marking.type in restricted_markings:
                done = True
                penalty = -320
                # print('l')
                return penalty, done  # Penalize and stop checking further
            if marking.type in soft_restricted_markings:
                done = True
                penalty = -160
                # print('sl')
                return penalty, done  # Penalize and stop checking further

    return penalty, done  # No penalty if only permissible markings were crossed


def calculate_steer_change_penalty(steer, previous_steer, weight=50.0):
    """Quadratic penalty"""
    steer_change = abs(steer - previous_steer)
    return -(steer_change ** 2) * weight


def calculate_oscillation_penalty(steering_history):
    """
    Penalize rapid left-right-left oscillations
    Returns penalty (negative value)
    """
    if len(steering_history) < 3:
        return 0.0
    
    # Check for sign changes (crossing zero)
    sign_changes = 0
    for i in range(len(steering_history) - 1):
        if steering_history[i] * steering_history[i+1] < 0:  # Different signs
            sign_changes += 1
    
    # Penalize if too many sign changes (oscillating)
    if sign_changes >= 3:  # 3+ sign changes in 5 steps = oscillating
        return -20.0 # 20.0
    elif sign_changes == 2:
        return -5.0 # 5.0
    
    return 0.0

''' Penalties for deviate form the valid path '''

def calculate_penalty_angle(done, angle_difference):
    abgle_threshold = 1  # angle for no penalty
    angle_difference = abs(angle_difference) # ignor direction
    angle_penalty_weight = 30.0  # Weight for angle penalty
    penalty = 0
    
    if angle_difference > 18: # 18
        penalty = -300
        # print(f'a : {angle_difference}')
        done = True
    elif angle_difference > abgle_threshold:
        penalty = (angle_difference / 180) * angle_penalty_weight  # Normalize angle to a 0-1 scale
        penalty = -(penalty*penalty)
        
    return penalty,done

	
def calculate_penalty_distance(done, distance):
    # Define the threshold for minimal penalty
    distance_threshold = 0.5  # Distance in meters for no penalty
    distance_penalty_weight = 2.0  # Weight for distance penalty
    distance_penalty = 0
      
    if distance > 4:
        distance_penalty = -300
        done = True
        # print(f'd : {distance}')
    # Calculate distance penalty only if outside the threshold
    elif distance > distance_threshold:
        distance_penalty = (distance - distance_threshold) * distance_penalty_weight
        distance_penalty = -(distance_penalty*distance_penalty)
        
    return distance_penalty,done 
```

```python
'''
RL environment for training sctript
'''

import time
import math 
import cv2
import gym
import carla
import random
import numpy as np
from gym import spaces
from tensorflow.keras.models import load_model


HEIGHT  = 240
WIDTH   = 320

SPIN = 5 # angle of random spin
MAX_SPEED = 80    # 80kmh
SPEED_THRESHOLD = 2 # defines when we get close to desired speed so we drop the throttle
MAX_WP_DIST = 500 # 500m

HEIGHT_REQUIRED_PORTION = 0.5 # bottom share, e.g. 0.1 is take lowest 10% of rows
WIDTH_REQUIRED_PORTION  = 0.9
CAMERA_POS_Z = 1.3 
CAMERA_POS_X = 1.4

SHOW_PREVIEW = True  # simulator veiw
SHOW_CAM     = False # seg camera veiw
MAX_STEPS    = 4090
PREFERRED_SPEED = 40

WP_TYPE_MAPPING = {
		0: [1, 0],  # normal
		1: [0, 1],  # destination
}

STEERING_VALUES = (-0.9, -0.75, -0.5, -0.25, -0.1, -0.05, 0.0, 0.05, 0.1, 0.25, 0.5, 0.75 ,0.9)

CNN_MODEL_PATH = 'cnn/cnn_fe_model.h5' # CNN feature extraction model path


class CarEnv(gym.Env):
	''' 
    Environment Initialization:
    
    CarEnv class inherits from gym.Env and initializes the environment with the CARLA client, vehicle model, and camera setup.
    camera provides semantic segmentation images to help the RL agent understand its environment visually,
    CNN model loaded from a pre-trained Keras model processes the camera input to generate the observation for the agent.
    '''
	
	def __init__(self, env_map=None):
		super(CarEnv, self).__init__()
  
		self.sensor_list = []  # Store sensor objects
		self.front_camera = None

		# Termination tracking
		self.termination_reason = None
		self.episode_terminations = {
			'collision': 0,
			'lane_invasion': 0,
			'angle_deviation': 0,
			'distance_deviation': 0,
			'destination_reached': 0,
			'max_steps': 0
		}
  
        # Define action and observation space
        # They must be gym.spaces objects
		self.action_space = spaces.MultiDiscrete([11])

		self.height_from = int(HEIGHT * (1 -HEIGHT_REQUIRED_PORTION))
		self.width_from  = int((WIDTH - WIDTH * WIDTH_REQUIRED_PORTION) / 2)
		self.width_to    = self.width_from + int(WIDTH_REQUIRED_PORTION * WIDTH)

		self.observation_space = spaces.Dict({
			'img' : spaces.Box(low=0.0, high=1.0, shape=(7, 18, 8), dtype=np.float32),  # Segmentation image
			'cspd': spaces.Box(low=0.0, high=1.0, shape=(1,), dtype=np.float32),        # Current speed
			'steer_hist': spaces.Box(low=-1.0, high=1.0, shape=(5,), dtype=np.float32), # Last 5 steering values
			'dist': spaces.Box(low=0.0, high=1.0, shape=(1,), dtype=np.float32),        # Normalized distance
			'ang' : spaces.Box(low=-1.0, high=1.0, shape=(2,), dtype=np.float32),       # Angle
			'typ': spaces.MultiBinary(2)                                                # Waypoint type
		})
		
		self.current_step = 0  # Current step count
		self.steering_history = [0.0] * 5  # Last 5 steering values

		self.client = carla.Client("localhost", 2000)
		self.client.set_timeout(30.0)
		self.world = self.client.get_world()
  
		if env_map is not None:
			print("Loading test map...")
			self.client.load_world(env_map)
		
		self.map = self.world.get_map()
		self.settings = self.world.get_settings()
		self.settings.no_rendering_mode = not SHOW_PREVIEW
		self.world.apply_settings(self.settings)
  
		self.spawn_points = self.map.get_spawn_points() 
		self.blueprint_library = self.world.get_blueprint_library()
		self.model_3 = self.blueprint_library.filter("model3")[0]
		self.cnn_model = load_model(CNN_MODEL_PATH,compile=False)
		self.cnn_model.compile()
		if SHOW_PREVIEW:
			self.spectator = self.world.get_spectator()
   
       
	def reset(self):
		''' 
        reinitializes the environment.
        
        '''
		self.cleanup() # clean env

		# 1. Reset all logic variables
		self.current_step = 0
		self.termination_reason = None
		self.steering_history = [0.0] * 5
		self.collision_hist = []
		self.lane_invade_hist = []
		self.front_camera = None
  
  
		# 2. Spawn vehicle and route
		# Check if we're using predefined segments (for testing)
		if hasattr(self, 'test_route'): # For test
			self.route = self.test_route
			spawn_point = self.spawn_points[0]
		elif hasattr(self, 'routes'):
			# indices = [4,13,15]
			# random_index = random.choice(indices)
			# self.route = self.routes[15]
			self.route = random.choice(self.routes)
			spawn_point = self.spawn_points[0]
		else: # Normal train
			# Use original path planner (for training)
			self.route, spawn_point = path_planner(self.spawn_points, self.map)
  
  
		self.vehicle = self.world.spawn_actor(self.model_3, spawn_point)
		# route wont start with the passed start point (spawn point) by default GRP in Carla
		# so we have to move the vehicle to starting point of the route 
		trans = self.route[0][0].transform

		# apply a random yaw adjustment
		trans.rotation.yaw += random.randrange(-SPIN, SPIN, 1)
		# Set the modified transform to the vehicle
		self.vehicle.set_transform(trans)
  
		# 3. Add a small wait and force a world tick
		for _ in range(10):
			self.world.tick()
  
		self.spectator.set_transform(carla.Transform(trans.location + carla.Location(z=20),carla.Rotation(yaw =-180, pitch=-90)))
		time.sleep(0.2)
  
		self.path_waypoints = get_path_waypoints(self.route) # path main wps
		currnet_wp_data = update_waypoint(self.path_waypoints, 0, trans.location)
		self.current_wp_idx = currnet_wp_data[1]
     
		# 4. Sensor setup
		self.sem_cam = self.blueprint_library.find('sensor.camera.semantic_segmentation')
		self.sem_cam.set_attribute("image_size_x", f"{WIDTH}")
		self.sem_cam.set_attribute("image_size_y", f"{HEIGHT}")
		self.sem_cam.set_attribute("fov", f"90")
		
		camera_init_trans = carla.Transform(carla.Location(z=CAMERA_POS_Z,x=CAMERA_POS_X))
		self.sensor = self.world.spawn_actor(self.sem_cam, camera_init_trans, attach_to=self.vehicle)
		self.sensor_list.append(self.sensor)
		self.sensor.listen(lambda data: self.process_img(data))

		self.vehicle.apply_control(carla.VehicleControl(throttle=0.0, brake=0.0))
		time.sleep(0.2)
  
		# showing camera at the spawn point
		if SHOW_CAM:
			cv2.namedWindow('Sem Camera',cv2.WINDOW_AUTOSIZE)
			cv2.imshow('Sem Camera', self.front_camera)
			cv2.waitKey(1)
						
		colsensor = self.blueprint_library.find("sensor.other.collision")
		self.colsensor = self.world.spawn_actor(colsensor, camera_init_trans, attach_to=self.vehicle)
		self.sensor_list.append(self.colsensor)
		self.colsensor.listen(lambda event: self.collision_data(event))

		lanesensor = self.blueprint_library.find("sensor.other.lane_invasion")
		self.lanesensor = self.world.spawn_actor(lanesensor, camera_init_trans, attach_to=self.vehicle)
		self.sensor_list.append(self.lanesensor)
		self.lanesensor.listen(lambda event: self.lane_data(event))

		while self.front_camera is None:
			time.sleep(0.1)
   
   
		# 5. First observations
		self.image_for_CNN = self.apply_cnn(self.front_camera[self.height_from:,self.width_from:self.width_to])
		self.cspd = 0.0
		self.dist = currnet_wp_data[4]
		self.ang = get_angle(trans.rotation.yaw, currnet_wp_data[0].transform.rotation.yaw)
		self.type = currnet_wp_data[2]
  
		self.start_dist = self.dist # copy start dist for reward calculations
  
		return self.create_observation()


	def create_observation(self):
		# Normalize distance
		norm_dist = min(self.dist / MAX_WP_DIST, 1.0)

		# Normalize current speed
		norm_cspd = min(self.cspd / MAX_SPEED, 1.0)

		# Normalize angle using sine and cosine
		norm_angle_sin = np.sin(np.radians(self.ang))
		norm_angle_cos = np.cos(np.radians(self.ang))
		norm_angle = [norm_angle_sin, norm_angle_cos]

		# One-hot encode waypoint type
		wp_type_ohe = WP_TYPE_MAPPING[self.type]
  
		observation = {
			'img' : self.image_for_CNN,                       # Segmentation image
			'cspd': np.array([norm_cspd], dtype=np.float32),  # Current speed
   			'steer_hist': np.array(self.steering_history, dtype=np.float32),  # Last 5 steers
			'dist': np.array([norm_dist], dtype=np.float32),  # Distance
			'ang' : np.array(norm_angle, dtype=np.float32),   # Angle
			'typ' : np.array(wp_type_ohe, dtype=np.float32)   # Waypoint type
		}
  
		return observation


	def step(self, action):
     
		'''
		step function maps actions to steering angles and applies vehicle control for throttle and steering.
			
		rewards are calculated based on distance traveled, collisions, lane invasions, and steering lock, which 
		penalizes prolonged sharp turns.

		'''
		trans = self.vehicle.get_transform()
		if SHOW_PREVIEW:
			self.spectator.set_transform(carla.Transform(trans.location + carla.Location(z=20),carla.Rotation(yaw =-180, pitch=-90)))

		# Extracting actions
		steer = STEERING_VALUES[action[0]]
  
		# Calculate steering change penalty using most recent steer (last in history)
		previous_steer = self.steering_history[-1]
		steer_change_penalty = calculate_steer_change_penalty(steer, previous_steer)
		
		# Update steering history (shift left and add new value)
		self.steering_history = self.steering_history[1:] + [steer]
		
		# Get the current speed of the vehicle
		velocity = self.vehicle.get_velocity()
		self.cspd = int(3.6 * math.sqrt(velocity.x**2 + velocity.y**2 + velocity.z**2))
		estimated_throttle = self.maintain_speed(self.cspd)
		# map throttle and apply steer and throttle	
		self.vehicle.apply_control(carla.VehicleControl(throttle=estimated_throttle, steer=steer, brake = 0.0))

		# storing camera to return at the end in case the clean-up function destroys it
		cam = self.front_camera
		# showing image
		if SHOW_CAM:
			cv2.imshow('Sem Camera', cam)
			cv2.waitKey(1)
		
		# start defining reward from each step
		reward = 0
		done = False # conditional fail
		truncated = False # max step reached
		
		# Increment the step counter
		self.current_step += 1 
		
		current_loc = self.vehicle.get_location()
		currnet_wp_data = update_waypoint(self.path_waypoints, self.current_wp_idx, current_loc)
		self.current_wp_idx = currnet_wp_data[1]
		if currnet_wp_data[3]:
			self.start_dist = currnet_wp_data[4]
		
		vehicle_yaw = trans.rotation.yaw
		dist_and_angle = get_dis_ang_from_nearest_wp(current_loc, vehicle_yaw, self.route)
		
		# Calculate all rewards and penalties first
		reward += calculate_distance_to_waypoint_reward(currnet_wp_data[3], currnet_wp_data[4], self.start_dist)
		
		penalty_1, done_angle = calculate_penalty_angle(False, dist_and_angle[1])
		penalty_2, done_dist = calculate_penalty_distance(False, dist_and_angle[0])
		penalty_3, done_collision = collision_penalty(False, self.collision_hist)
		penalty_4, done_lane = lane_invasion_penalty(False, self.lane_invade_hist)
		
		oscillation_penalty = calculate_oscillation_penalty(self.steering_history)
		reward += (penalty_1 + penalty_2 + penalty_3 + penalty_4 + steer_change_penalty + oscillation_penalty)
		
		# NOW check termination conditions in priority order
		# Priority 1: Collision (most critical)
		if done_collision:
			done = True
			self.termination_reason = 'collision'
		# Priority 2: Distance deviation (off track)
		elif done_dist:
			done = True
			self.termination_reason = 'distance_deviation'
		# Priority 3: Angle deviation
		elif done_angle:
			done = True
			self.termination_reason = 'angle_deviation'
		# Priority 4: Lane invasion
		elif done_lane:
			done = True
			self.termination_reason = 'lane_invasion'
		# Priority 5: Destination reached (success!)
		elif currnet_wp_data[5]:
			reward += 1000 # 500 Normal train, 1000 fine-tune
			done = True
			self.termination_reason = 'destination_reached'
		# Priority 6: Max steps (timeout)
		elif self.current_step >= MAX_STEPS:
			print("Step")
			done = True
			truncated = True
			self.termination_reason = 'max_steps'

		# Create info dict
		info = {
			'truncated': truncated,
		}
  
		if done:
			# Add termination reason to info
			if self.termination_reason:
				info['termination_reason'] = self.termination_reason
			
			# Return current observation (no need to compute next state if done)
			observation = self.create_observation()
			return observation, reward, done, info
		
		# Only compute next state if NOT done
		self.image_for_CNN = self.apply_cnn(self.front_camera[self.height_from:,self.width_from:self.width_to])
		self.dist = currnet_wp_data[4]
		self.ang = get_angle(vehicle_yaw, currnet_wp_data[0].transform.rotation.yaw)
		self.type = currnet_wp_data[2]
		
		observation = self.create_observation()

		return observation, reward, done, info

	
	def maintain_speed(self,s):
		'''
		this is a very simple function to maintan desired speed
		s arg is actual current speed
		'''
		if s >= PREFERRED_SPEED:
			return 0
		elif s < PREFERRED_SPEED - PREFERRED_SPEED:
			return 0.7 # think of it as % of "full gas"
		else:
			return 0.3 # tweak this if the car is way over or under preferred speed 


	def cleanup(self):
		''' 
        function to cealn carla env
        destroy all the actors in the env
        '''
        # Stop sensor streams before destroying them
		for sensor in self.sensor_list:
			if sensor.is_alive:
				sensor.stop() 
				sensor.destroy()
		self.sensor_list.clear()
                
		actors = self.world.get_actors().filter('*vehicle*')
		for vehicle in actors:
			vehicle.destroy()

		# Tick a few times to ensure the server registers the deletions
		for _ in range(5):
			self.world.tick()


	def apply_cnn(self,im):
		''' 
        apply_cnn
        '''
		img = np.float32(im)
		img = img /255
		img = np.expand_dims(img, axis=0)
		cnn_applied = self.cnn_model([img,0],training=False)
		cnn_applied = np.squeeze(cnn_applied)
		return  cnn_applied ##[0][0]

 
	def process_img(self, image):
		image.convert(carla.ColorConverter.CityScapesPalette)
		i = np.array(image.raw_data)
		i = i.reshape((HEIGHT, WIDTH, 4))[:, :, :3] # this is to ignore the 4th Alpha channel - up to 3
		self.front_camera = i


	def collision_data(self, event):
		self.collision_hist.append(event)


	def lane_data(self, event):
		self.lane_invade_hist.append(event)
```

#### **5.4. Custom Logs:**

```python
from stable_baselines3.common.callbacks import BaseCallback


class TerminationLoggingCallback(BaseCallback):
    def __init__(self, verbose=0):
        super(TerminationLoggingCallback, self).__init__(verbose)
         
        # Counters for current iteration
        self.iteration_termination_counts = {
            'collision': 0,
            'lane_invasion': 0,
            'angle_deviation': 0,
            'distance_deviation': 0,
            'destination_reached': 0,
            'max_steps': 0
        }
        
        self.iteration_episodes = 0
        
    def _on_rollout_start(self) -> None:
        """Reset counters at the start of each iteration"""
        self.iteration_termination_counts = {
            'collision': 0,
            'lane_invasion': 0,
            'angle_deviation': 0,
            'distance_deviation': 0,
            'destination_reached': 0,
            'max_steps': 0
        }
        self.iteration_episodes = 0
        
    def _on_step(self) -> bool:
        if len(self.locals['infos']) > 0:
            info = self.locals['infos'][0]
            
            if 'episode' in info:
                self.iteration_episodes += 1
                
                # Get termination reason from environment
                termination = info['termination_reason']
                
                # Get episode reward
                episode_reward = info['episode']['r']
                
                # Increment counter
                if termination in self.iteration_termination_counts:
                    self.iteration_termination_counts[termination] += 1
                
                # Log to console
                if self.verbose > 0:
                    print(f"Episode {self.iteration_episodes} ended: {termination}, Reward: {episode_reward:.2f}")
        
        return True
    
    def _on_rollout_end(self) -> None:
        env = self.model.get_env().envs[0]
        env.unwrapped.vehicle.apply_control(carla.VehicleControl(throttle=0.0, steer=0.0, brake=1.0))
        print("Stop vehicle after rollout.")
        
        """Log iteration statistics"""
        # Log counts
        for reason, count in self.iteration_termination_counts.items():
            self.logger.record(f'termination/count_{reason}', count)

        self.logger.record('termination/total_episodes', self.iteration_episodes)
```

```python
import tensorflow as tf
from tensorflow.python.summary.summary_iterator import summary_iterator
import matplotlib.pyplot as plt
import pandas as pd
import glob


def read_tensorboard_logs(log_dir, tag=None):
    """Read scalar data from TensorBoard logs"""
    data = []
    tags_found = set()
    
    # Find all event files recursively
    event_files = glob.glob(f"{log_dir}/**/events.out.tfevents.*", recursive=True)
    
    print(f"Reading {len(event_files)} event files...")
    
    for event_file in event_files:
        try:
            for event in summary_iterator(event_file):
                for value in event.summary.value:
                    tags_found.add(value.tag)
                    if tag is None or value.tag == tag:
                        data.append({
                            'step': event.step,
                            'value': value.simple_value,
                            'tag': value.tag,
                            'wall_time': event.wall_time
                        })
        except Exception as e:
            print(f"Error reading {event_file}: {e}")
            
    print(f"\nAvailable tags ({len(tags)}):")
    for tag in tags:
        print(f"  - {tag}")
    
    return pd.DataFrame(data), sorted(tags_found)


def plot_metric(df, metric_name, figsize=(8, 4), y_label='Value', color='#ff8100'):
    """
    Plot a single metric from TensorBoard logs
    
    Args:
        df: DataFrame containing the TensorBoard data
        metric_name: Name of the metric to plot (e.g., 'rollout/ep_rew_mean')
        figsize: Tuple of (width, height) for the figure
        color: Color for the line plot
        show_stats: Whether to print statistics
    """
    # Filter data for the specific metric
    metric_df = df[df['tag'] == metric_name].sort_values('step')
    
    if metric_df.empty:
        print(f"Error: Metric '{metric_name}' not found in data")
        return
    
    # Create plot
    plt.figure(figsize=figsize)
    plt.plot(metric_df['step'], metric_df['value'], color=color)
    plt.xlabel('Training Step', fontsize=10)
    plt.ylabel(y_label, fontsize=10)
    plt.title(metric_name, fontsize=10, fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Print statistics
    print(f"\n{metric_name} Statistics:")
    print(f"  Min:   {metric_df['value'].min():.4f}")
    print(f"  Max:   {metric_df['value'].max():.4f}")
    print(f"  Mean:  {metric_df['value'].mean():.4f}")
    print(f"  Final: {metric_df['value'].iloc[-1]:.4f}")
```

Test Method and Utility Fuctions

```python
'''
5.1, 5.2, 5.3
'''

import os
import csv
from stable_baselines3 import PPO
from collections import defaultdict


# Load the trained model
iteration = 75 # Iteration of the saved model that need to be loaded!!
num_episodes = 20 # Number of test episodes
model_path = f"ppo/models/phase2/ppo_model_iteration_{iteration}.zip"


def save_results_to_file(iteration, episode_results, test_map=None):
    # Save results to file
    results_dir = "ppo/results/data"
    os.makedirs(results_dir, exist_ok=True)
    
    if test_map is None:
        test_map = "town10"

    # Save detailed results
    csv_path = os.path.join(results_dir, f"test_results_of_model_{iteration}_for_{test_map}.csv")
    with open(csv_path, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['episode', 'reward', 'termination', 'steps'])
        writer.writeheader()
        writer.writerows(episode_results)

    print(f"\nDetailed results saved to: {csv_path}")


def test_model(test_map=None, num_episodes=20):
    # Load the environment
    env = CarEnv(env_map=test_map)
    model = PPO.load(model_path, env=env)

    # Get the top 20 longest road segments
    routes = find_longest_non_junction_paths(env.map, top_n=num_episodes)

    # Storage for results
    episode_rewards = []
    termination_counts = defaultdict(int)
    episode_results = []

    for episode_idx, route in enumerate(routes):
        env.test_route = route
        obs = env.reset()
        done = False
        episode_reward = 0
        step_count = 0
        
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, info = env.step(action)
            episode_reward += reward
            step_count += 1
            
            if done:
                termination = info.get('termination_reason', 'unknown')
                termination_counts[termination] += 1
                
                episode_results.append({
                    'episode': episode_idx + 1,
                    'reward': episode_reward,
                    'termination': termination,
                    'steps': step_count
                })
                
                episode_rewards.append(episode_reward)
                print(f"Episode {episode_idx + 1}/{num_episodes} | Reward: {episode_reward:.2f} | Steps: {step_count} | Termination: {termination}")
                
    save_results_to_file(iteration, episode_results, test_map)
    print("\nEvaluation complete!")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
import numpy as np


def visualize_test_results(csv_path, map_name="Test Results", save_dir=None):
    
    # Set style
    sns.set_style("whitegrid")
    plt.rcParams['figure.figsize'] = (15, 10)
    
    # Load data
    df = pd.read_csv(csv_path)
    df['reward_per_step'] = df['reward'] / df['steps']
    
    # Create figure with subplots
    fig, axes = plt.subplots(2, 2, figsize=(18, 10))
    fig.suptitle(f'Test Results Analysis - {map_name}', fontsize=16, fontweight='bold')
    plt.subplots_adjust(hspace=0.3)
    
    # 1. Success Rate Pie Chart
    ax1 = axes[0, 0]
    success_counts = df['termination'].value_counts()
    colors = ['#73eda6' if 'destination' in str(idx) else '#eb6a5c' for idx in success_counts.index]
    ax1.pie(success_counts.values, labels=success_counts.index, autopct='%1.1f%%', 
            colors=colors, startangle=90, textprops={'fontsize': 12})
    ax1.set_title('Termination Reasons Distribution', fontweight='bold')
    
    # 2. Reward Distribution
    ax2 = axes[0, 1]
    ax2.hist(df['reward'], bins=20, color='#3498db', alpha=0.7)
    ax2.axvline(df['reward'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {df["reward"].mean():.2f}')
    ax2.set_xlabel('Reward', fontweight='bold')
    ax2.set_ylabel('Frequency', fontweight='bold')
    ax2.set_title('Reward Distribution', fontweight='bold')
    ax2.legend()
    ax2.grid(axis='y', alpha=0.3)
    
    # 3. Episode Length (Steps) Distribution
    ax3 = axes[1, 0]
    ax3.hist(df['steps'], bins=20, color='#f39c12', alpha=0.7)
    ax3.axvline(df['steps'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {df["steps"].mean():.1f}')
    ax3.set_xlabel('Steps', fontweight='bold')
    ax3.set_ylabel('Frequency', fontweight='bold')
    ax3.set_title('Episode Length Distribution', fontweight='bold')
    ax3.legend()
    ax3.grid(axis='y', alpha=0.3)
    
    # 4. Reward per Step Distribution
    ax4 = axes[1, 1]
    ax4.hist(df['reward_per_step'], bins=20, color='#9b59b6', alpha=0.7)
    ax4.axvline(df['reward_per_step'].mean(), color='red', linestyle='--', linewidth=2, 
                label=f'Mean: {df["reward_per_step"].mean():.4f}')
    ax4.set_xlabel('Reward per Step', fontweight='bold')
    ax4.set_ylabel('Frequency', fontweight='bold')
    ax4.set_title('Reward Efficiency Distribution', fontweight='bold')
    ax4.legend()
    ax4.grid(axis='y', alpha=0.3)
    
    # Save if directory provided
    if save_dir:
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)
        save_path = os.path.join(save_dir, f"{map_name.replace(' ', '_')}_analysis.png")
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Plot saved to: {save_path}")
    
    plt.show()


def compare_maps_visual(csv_path1, csv_path2, map1_name="Map 1", map2_name="Map 2", save_dir=None):
    
    # Load data
    df1 = pd.read_csv(csv_path1)
    df2 = pd.read_csv(csv_path2)
    
    df1['reward_per_step'] = df1['reward'] / df1['steps']
    df2['reward_per_step'] = df2['reward'] / df2['steps']
    
    df1['map'] = map1_name
    df2['map'] = map2_name
    
    # Combine for some plots
    df_combined = pd.concat([df1, df2])
    
    # Create figure
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle(f'Comparison: {map1_name} vs {map2_name}', fontsize=16, fontweight='bold', y=1.02)
    plt.subplots_adjust(wspace=0.3)
    
    # 1. Success Rate Comparison (Bar Chart)
    ax1 = axes[0]
    success_rate1 = (df1['termination'] == 'destination_reached').sum() / len(df1) * 100
    success_rate2 = (df2['termination'] == 'destination_reached').sum() / len(df2) * 100
    
    bars = ax1.bar([map1_name, map2_name], [success_rate1, success_rate2], 
                   color=['#3498db', '#e74c3c'], alpha=0.7)
    ax1.set_ylabel('Success Rate (%)', fontweight='bold')
    ax1.set_title('Success Rate Comparison', fontweight='bold', pad=20)
    ax1.set_ylim(0, 100)
    
    # Add value labels on bars
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.1f}%', ha='center', va='bottom', fontweight='bold')
    ax1.grid(axis='y', alpha=0.3)
    
    # 2. Termination Reasons Comparison (Stacked Bar)
    ax2 = axes[1]
    term_counts1 = df1['termination'].value_counts()
    term_counts2 = df2['termination'].value_counts()
    
    all_terms = list(set(term_counts1.index) | set(term_counts2.index))
    map1_percentages = [(term_counts1.get(term, 0) / len(df1) * 100) for term in all_terms]
    map2_percentages = [(term_counts2.get(term, 0) / len(df2) * 100) for term in all_terms]
    
    x_pos = [0, 1]
    bottom1 = 0
    bottom2 = 0
    
    colors_list = plt.cm.Set3(np.linspace(0, 1, len(all_terms)))
    
    for i, term in enumerate(all_terms):
        ax2.bar(0, map1_percentages[i], bottom=bottom1, label=term if i == 0 else "", 
                color=colors_list[i])
        ax2.bar(1, map2_percentages[i], bottom=bottom2, color=colors_list[i])
        bottom1 += map1_percentages[i]
        bottom2 += map2_percentages[i]
    
    ax2.set_ylabel('Percentage (%)', fontweight='bold')
    ax2.set_title('Termination Reasons Distribution', fontweight='bold', pad=20)
    ax2.set_xticks([0, 1])
    ax2.set_xticklabels([map1_name, map2_name])
    ax2.set_ylim(0, 100)
    
    # Create custom legend
    legend_elements = [Patch(facecolor=colors_list[i], label=term) for i, term in enumerate(all_terms)]
    ax2.legend(handles=legend_elements, loc='lower left', bbox_to_anchor=(0, 0), fontsize=10)
    ax2.grid(axis='y', alpha=0.3)
    
    # 3. Reward per Step Comparison
    ax3 = axes[2]
    avg_reward1 = df1['reward_per_step'].mean()
    avg_reward2 = df2['reward_per_step'].mean()

    # Bar chart
    bars = ax3.bar([map1_name, map2_name], [avg_reward1, avg_reward2], color=['#3498db', '#e74c3c'], alpha=0.7)

    # Add labels and title
    ax3.set_ylabel('Average Reward per Step', fontweight='bold')
    ax3.set_title('Reward Efficiency Comparison', fontweight='bold', pad=20)
    ax3.grid(axis='y', alpha=0.3)
    
    # Add value labels on bars
    for bar in bars:
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.4f}', ha='center', va='bottom', fontweight='bold')
    ax3.grid(axis='y', alpha=0.3)
    
    # Save if directory provided
    if save_dir:
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)
        save_path = os.path.join(save_dir, "comparison_analysis.png")
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Comparison plot saved to: {save_path}")
    
    plt.show()
```

```python
'''
5.1, 5.2, 5.3
'''

import os
import csv
from stable_baselines3 import PPO
from collections import defaultdict


# Load the trained model
iteration = 75 # Iteration of the saved model that need to be loaded!!
num_episodes = 20 # Number of test episodes
model_path = f"ppo/models/phase2/ppo_model_iteration_{iteration}.zip"


def save_results_to_file(iteration, episode_results):
    # Save results to file
    results_dir = "ppo/results/data"
    os.makedirs(results_dir, exist_ok=True)
        
    fieldnames = episode_results[0].keys() if episode_results else []

    # Save detailed results
    csv_path = os.path.join(results_dir, f"test_results_of_model_{iteration}_in_weathers.csv")
    with open(csv_path, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(episode_results)

    print(f"\nDetailed results saved to: {csv_path}")
    

def randomize_weather(world, index=None):
    # Available weather presets in Carla (carla.WeatherParameters)
    WEATHER_PRESETS = [
        carla.WeatherParameters.MidRainSunset,
        carla.WeatherParameters.HardRainSunset,
    ]

    # Define custom weather settings (cloudiness, precipitation, etc.)
    CUSTOM_WEATHER_1 = carla.WeatherParameters(
        cloudiness=100.0,  # 0 to 100 (0: clear sky, 100: overcast)
        precipitation=100.0,  # 0 to 100 (0: no precipitation, 100: heavy rain)
        precipitation_deposits=100.0,  # 0 to 100 (0: no water, 100: wet ground)
        wind_intensity=100.0,  # 0 to 100 (0: no wind, 100: strong wind)
        sun_azimuth_angle=180.0,  # 0 to 360 (angle of sun)
        sun_altitude_angle=-90.0,  # -90 to 90 (angle of the sun in the sky)
        fog_density = 100 # 0 to 100 (how dense the fog is in the atmosphere)
    )

    CUSTOM_WEATHER_2 = carla.WeatherParameters(
        cloudiness=90.0,
        precipitation=0.0,
        precipitation_deposits=70.0,
        wind_intensity=100.0,
        sun_azimuth_angle=180.0,
        sun_altitude_angle=-50.0,
        fog_density=40.0
    )

    CUSTOM_WEATHER_3 = carla.WeatherParameters(
        cloudiness= 0.0,
        precipitation=10.0,
        precipitation_deposits=10.0,
        wind_intensity=50.0,
        sun_azimuth_angle=180.0,
        sun_altitude_angle=90.0,
        fog_density = 20
    )

    # Add custom weather to the list
    CUSTOM_WEATHERS = [CUSTOM_WEATHER_1, CUSTOM_WEATHER_2, CUSTOM_WEATHER_3]
    
    # Combine both default and custom weather settings into a dictionary
    WEATHER_OPTIONS = {
        'MidRainSunset': WEATHER_PRESETS[0],
        'HardRainSunset': WEATHER_PRESETS[1],
        'SevereStorm': CUSTOM_WEATHER_1,
        'ClearWithFog': CUSTOM_WEATHER_2,
        'BlindingSun': CUSTOM_WEATHER_3,
    }

    selected_weather_name, selected_weather = random.choice(list(WEATHER_OPTIONS.items()))
    world.set_weather(selected_weather)

    return selected_weather_name


def test_model_for_weather(test_map=None, num_episodes=20):
    # Load the environment
    env = CarEnv(env_map=test_map)
    model = PPO.load(model_path, env=env)

    # Get the top 20 longest road segments
    routes = find_longest_non_junction_paths(env.map, top_n=num_episodes)

    # Storage for results
    episode_rewards = []
    termination_counts = defaultdict(int)
    episode_results = []

    for episode_idx, route in enumerate(routes):
        
        weather = randomize_weather(env.world)
        
        env.test_route = route
        obs = env.reset()
        done = False
        episode_reward = 0
        step_count = 0
        
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, info = env.step(action)
            episode_reward += reward
            step_count += 1
            
            if done:
                termination = info.get('termination_reason', 'unknown')
                termination_counts[termination] += 1
                
                episode_results.append({
                    'episode': episode_idx + 1,
                    'weather': weather,
                    'reward': episode_reward,
                    'termination': termination,
                    'steps': step_count
                })
                
                episode_rewards.append(episode_reward)
                print(f"Episode {episode_idx + 1}/{num_episodes} | Weather: {weather} | Reward: {episode_reward:.2f} | Steps: {step_count} | Termination: {termination}")
                
    save_results_to_file(iteration, episode_results)
    print("\nEvaluation complete!")
```
